```python
def roll(current_player, round_score):
    dice_val = random.randint(1, 6)
    print(current_player + " decided to roll. " + current_player + 
          " rolled: " + str(dice_val))
    
    if (dice_val == 1):
        rolled_a_one = True;
        player_round_score = 0
        print("\n" + current_player + " rolled a 1 and lost the points for this round.")
        return -round_score
    return dice_val

def hold(current_player, round_score, total_score):
    print(current_player + " decided to hold.")
    return round_score + total_score

def under_winning_score():
    return player_total_score + player_round_score < winning_score and comp_total_score + comp_round_score < winning_score

def print_stats(current_player, round_score, total_score):
    print(current_player + " score this round: " + str(round_score))
    print(current_player + " total score: " + str(total_score + round_score))

import random

winning_score = 10

player_total_score = 0
player_round_score = 0

comp_total_score = 0
comp_round_score = 0

rolled_a_one = False

print("============================ RULES ============================")
print("Welcome to Pig!")
print("1. You will go first and roll a dice.")
print("2. Each number you roll will be added to the last number.")
print("3. If you roll a 1, though, your score for the round will")
print("   become 0!")
print("4. You can hold your points and not roll again for the round.")
print("Please enjoy the game!")
print("===============================================================\n")

while (under_winning_score()):
    
    #code for player's round
    ans = input("\nDo you want to roll or hold? ")
    
    while (ans == "roll" and not rolled_a_one and under_winning_score()):
        curr_value = roll("You", player_round_score)
        player_round_score += curr_value
        print_stats("Your", player_round_score, player_total_score)
        
        if (curr_value <= 0):
            rolled_a_one = True
            print("\nThe computer will go next. It's current score: " + str(comp_total_score))
        elif (under_winning_score()):
            ans = input("\nDo you want to roll or hold? ")

    if (ans == "hold"):
        player_total_score = hold("You", player_round_score, player_total_score)
        player_round_score = 0
        print("Your updated total score: " + str(player_total_score))
        print("\nThe computer will go next. It's current score: " + str(comp_total_score))
    rolled_a_one = False

    #code for computer's round
    choices = [1]
    choice = choices[random.randint(0, len(choices) - 1)]
    while (choice == 1 and not rolled_a_one and under_winning_score()):
        curr_value = roll("The computer", comp_round_score)
        comp_round_score += curr_value
        print_stats("The computer's", comp_round_score, comp_total_score)
        
        if (curr_value <= 0):
            rolled_a_one = True
            print("\nYou will go next. Your current score: " + str(player_total_score))
        elif (under_winning_score()):
            if (comp_round_score > 10):
                choices.append(2)
            choice = choices[random.randint(0, len(choices) - 1)]
            print("")

    if (choice == 2):
        comp_total_score = hold("The computer", comp_round_score, comp_total_score)
        comp_round_score = 0 
        print("The computer's updated total score: " + str(comp_total_score))
        print("\nYou will go next. Your current score: " + str(player_total_score))    
    rolled_a_one = False

player_total_score += player_round_score
comp_total_score += comp_round_score

if (player_total_score >= winning_score):
    print("\nYou won with " + str(player_total_score) + " points!")
elif (comp_total_score >= winning_score):
    print("\nThe computer won with " + str(comp_total_score) + " points!")    
```
